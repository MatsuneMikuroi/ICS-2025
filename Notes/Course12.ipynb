<u>Remarque</u>

Même si $\Phi(\alpha) = \alpha$, la converge de la suite de Picard n'est pas assurée.

<u>Théorème du point fixe de Bronwer</u>

Si $\Phi(x)$ est continue sur $[a, b]$ et si 
$$\forall x \in [a, b], \Phi(x) \in [a, b]$$
alors $\Phi(x)$ admet au moins un point fixe $\alpha$ sur $[a, b]$


<u>Preuve</u>

<span style="color:green"> Schéma </span>

Soit $d(x) = \Phi(x) - x$. On cherhe à montrer qu'il existe $\alpha \in [a, b]$ tel que $d(\alpha) = 0$

$$d(a) \cdot d(b) = \underset{\geq 0}{\underbrace{(\Phi(a) - a)}} \underset{\leq 0}{\underbrace{(\Phi(b) - b)}}$$
Ainsi, la fonction $d(x)$ change de signe sur l'intervalle $[a, b]$ et est continue. Par le théorèem de Bolzano, il existe $\alpha \in [a, b]$ tel que $d(\alpha) = 0$, donc $\Phi(\alpha) = \alpha$

<div style="text-align:right"> &#x220E; </div>


<u>Définition</u>

Une fonction $\Phi(x)$ est dite $K$-contracte sur un intervalle $[a, b] = I \subset \R$ si
$$\exist K \in \R, 0 < k < 1 | \forall x, y \in I, |\Phi(x) - \Phi(y)| \leq k \cdot |x - y|$$

<span style="color:green"> Schéma </span>

<u>Remarques</u>

1. En développant $\Phi$ au premier ordre autour de $y \in I$, on a:
$$\Phi(x) = \Phi(y) + f^{\prime}(\xi)(x-y), \xi \in [x, y] \implies |\Phi(x) - \Phi(y)| = \underset{\equiv K}{\underbrace{|f^{\prime}(\xi)|}}|(x-y)|$$
Pour prouver que $\Phi$ est $K$-contractante sur $I$, il suffit de montrer que
$$\max_{\xi \in I}(f^{\prime}) < 1$$

2. Si la fonction est $K$-contractante sur $I$, alors admet un unique point fixe.

<u>Théorème</u>

Si $\Phi(x)$ satisfait les condition du théorèem de Bronwer et qu'en plus elle est $K$-contractante, alors elle admet un unique point fixe $\alpha$ sur $I$.


<u>Preuve</u>

Admettons, par l'absurde, que la fonction ait deux fixes différents $\alpha_{1}$ et $\alpha_{2}$ sur $I$. Alors la $K$-contraction nous dit que:
$$|\underset{=\alpha_{1}}{\underbrace{f(\alpha_{1})}} - \underset{=\alpha_{2}}{\underbrace{f(\alpha_{2})}}| \leq K |\alpha_{1} - \alpha_{2}| \implies |\alpha_{1} - \alpha_{2}| \leq K |\alpha_{1} - \alpha_{2}| < |\alpha_{1} - \alpha_{2}|$$
Ce resultat est abusrbe, il existe donc qu'un seul point fixe $\alpha$.

<div style="text-align:right"> &#x220E; </div>


<u>Théorème</u>

Soit $\Phi(x) \colon [a, b] \to \R$ une fonction $K$-contractante. Si $\Phi([a, b]) \subset [a, b]$, alors la suite $\{ x_{k}\}$ définie par $x_{k+1} = \Phi(x_{k})$ est convergente, i.e,
 $$\Phi([a, b]) \subset [a, b], x_{k+1} := \Phi(x_{k}) \implies \lim_{k\to \infty} x_{k+1} = l, l \in \R$$


<u>Preuve</u>

1. La suite est bornée. Comme $x_{0} \in [a, b]$ et $x_{k+1} = \Phi(x_{k}) \in [a, b]$, alors $x_{k} \in [a, b], \forall k$. Elle est donc bornée.

2. La suite est de Cauchy.
$$|x_{k+1} - x_{k}| \equiv |\Phi(x_{k}) - \Phi(x_{k-1})| \leq K |x_{k} - x_{k-1}| \leq K^{k} | x_{1} - x_{0}|$$
Cette quantité tend vers 0 quand $k \to \infty$, mais cela ne suffit pas pour montrer qu'elle est de Cauchy. On considère deux rangs $m$ et $n, m > n$.
$$\begin{aligned}
    |x_{m} - x_{n}| &\leq |x_{m} - x_{m-1}| + |x_{m-1} - x_{m-2}| + \ldots + |x_{n+1} - x_{n}|\\
    &= \sum_{k=n}^{m-1} |x_{k+1} - x_{k}|\\
    &\leq \sum_{k=m}^{m-1} K^{k}|x_{1} - x_{0}|\\
    =& |x_{1} - x_{0}| \sum_{k=n}^{m-1} K^{k}
\end{aligned}$$
Or
$$\sum_{k=n}^{m-1} K^{k} = K^{n} \frac{1-K^{m-n}}{1-K} < \frac{K^{n}}{1 - K} \to 0$$
Comme la suite est bornée et de Cauchy, alors elle converge.

<div style="text-align:right"> &#x220E; </div>


<u>Théorème</u>

Si la suite $\{x_{k}\}$ définie par l'itération de Picard est convergente vers $\overline{x}$ et si $\Phi$ est continue, alors $\overline{x}$ est un point fixe de $\Phi$.


<u>Preuve</u>

$$\overline{x} = \lim_{k \to \infty} x_{k+1} = \lim_{k \to \infty} \Phi(x_{k}) \overset{\text{continue}}{=} \Phi(\lim_{k \to \infty} x_{k}) = \Phi(\overline{x})$$

On a donc prouvé que soit $x_{0}$ et la suite $x_{k+1} = \Phi(x_{k}), \forall k > 0$, si
1. $\forall x \in [a, b], \Phi(x) \in [a, b]$
2. $\exist K, 0 < K < 1 | \forall x \in [a, b], |\Phi^{\prime}(x)| < K$
3. $\Phi(x) \in C^{0}([a, b])$

alors $\Phi(x)$ a un unique point fixe $\alpha$ dans $[a, b]$ et la suite $\{x_{k}\}$ converge vers $\alpha$ pour tout choix de $x_{0} \in [a, b]$.

<div style="text-align:right"> &#x220E; </div>


### Implémentation de la fonction d'itération de Picard

````python
# recherche du point fixe à l'aide de la méthode de Picard
def resoudre_par_Picard(phi,x_0,eps,k_max,couleur): 
    iteration = 0
    condition = True
    plt.plot([x_0], [x_0], color = couleur, marker='o')
    while iteration < k_max and condition:
        x_1 = phi(x_0)
        condition = abs(x_1-x_0) > eps
        plt.plot([x_0,x_0], [x_0,x_1], color = couleur)
        plt.plot([x_0,x_1], [x_1,x_1], color = couleur)
        x_0 = x_1
        iteration += 1
    return x_0,iteration
````

# Intégration numérique

## Introduction

Pour calculer une intégrale, on utilise le théorème fondamental du calcul intégral
$$\int_{a}^{b} f(x) \mathrm{d} x = F(b) - F(a)$$
où $F(x)$ est la primitive de $f(x)$.

<span style="color:gray"> exemple: 

$$\textcolor{gray}{\int_{0}^{1} x^{\prime} \mathrm{d} x= \left. \frac{1}{3} x^{3} \right|_{x = 1} - \left. \frac{1}{3} x^{3} \right|_{x = 0} = \frac{1}{3}}$$

</span>


<u>Définition du problème</u>

Soit $f \colon [a, b] \to \R$ une fonction continue. On cherche à approximer numériquement la quantité $\int_{a}^{b} f(x) \mathrm{d} x$. 

On se base sur des évaluations ponctuelles de $f$ en certains points $x_{i}$ stratégiquement choisis. On définit alors une méthode de quadrature comme la quantité
$$J(f) = \sum_{i} \omega_{i} f(x_{i}) \approx \int_{a}^{b} f(x) \mathrm{d} x$$



<u>Définition</u>

Une partition, aussi appelée discrétisation, de l'intervalle $[a, b]$ est un sous-ensemble de $N+1$ points $x_{0}, x_{1}, \ldots, x_{n}$ tels que
$$a = x_{0} < x_{1} < \ldots < x_{n} = b$$

<span style="color:green">Schéma</span>


Le nombre réel $h = \max|x_{i+1} - x_{i}|$ caractérise la finesse, aussi appelée le pas, de la partition.

En générale, on pose $h = \frac{b - a}{N}$ et $x_{i} = a + i \cdot h, i = 0, 1, \ldots, N$. On parle alors d'une partition régulière de $[a, b]$ avec un pas de $h$.


<u>Définition</u>

- Dans une méthode de quadrature non-composite, aussi dite simple, on approxime l'intégrale sur tout l'intervalle $[a, b]$.

<span style="color:green">Schéma</span>

- Dans une méthode composite, on commence par créer une partition de $[a, b]$, telle que
$$\int_{a}^{b} f(x) \mathrm{d} x = \sum_{i=0}^{n} \int_{x_{i}}^{x_{i+1}} f(x) \mathrm{d} x$$
<span style="color:green">Schéma</span>

et on cherche à approximer les intégrale 
$$\int_{x_{i}}^{x_{i+1}} f(x) \mathrm{d} x$$
par des formules de quadrature simple.

<span style="color:gray">

- Méthode à 1 point: $A = (b - a) \cdot f(\xi)$
    - $\xi = a$ (point de gauche)
    - $\xi = b$ (point de droite)
    - $\xi = \frac{b+a}{2}$ (point milieu)
    - $\xi \in$ ```random(a, b)``` (Riemann)
- Méthode à 2 points:
    - $\xi_{1} = a, \xi_{b} = b$ (trapèze)

</span>


## Interpolation de Lagrange

On veut trouver un polynôme $p(x)$ de degré $n$ qui interpole un fonction $f(x)$ en $n+1$ points.

Autrement dit, il prend des valeurs définies $p_{0}, p_{1}, \ldots, p_{n}$ en $x_{0}, x_{1}, \ldots, x_{n}$, i.e., 
$$f(x_{j}) = p_{j} = p(x_{j}), \forall j \in 0, 1, \ldots, n$$
On peut partir de
$$p(x) = a_{0} + a_{1} \cdot x + a_{2} \cdot x + \ldots + a_{n} x^{n}$$
et
$$\begin{aligned}
    p(x_{1}) &= f(x_{1})\\
    p(x_{2}) &= f(x_{2})\\
    \vdots\\
    p(x_{n}) &= f(x_{n})
\end{aligned}$$
On simplifie le problème en cherchant le polynôme $\varphi_{k}(x)$ tel que:
$$\begin{aligned}
    \varphi_{k}(x_{k}) = 1\\
    \varphi_{k}(x_{j}) = 0
\end{aligned}, \forall j \neq k$$

<span style="color:green">Schéma</span>

Le polynôme suivant satisfait ce critère:
$$\varphi_{k}(x) = \frac{(x - x_{0}) (x - x_{1}) \ldots (x - x_{k-1}) (x - x_{k+1}) \ldots (x - X_{n})}{(x_{k} - x_{0}) (x_{k} - x_{1}) \ldots (x_{k} - x_{k-1}) (x_{k} - x_{k+1}) \ldots (x_{k} - x_{n})} = \prod_{\begin{aligned} j = 0 \\ j \neq k \end{aligned}}^{n} \frac{x - x_{j}}{x_{k} - x_{j}}$$
On peut montrer que les polynômes $\{ \varphi_{0}, \ldots, \varphi_{n}\}$ forment une base de $\mathbb{P}_{n}$ appelée base de Lagrange associée aux points $x_{0}, x_{1}, \ldots, x_{n}$.

Ainsi, le polynôme $p(x)$ cherché est unique et vérifie 
$$p(x) = p_{0} \varphi_{0}(x) + p_{1} \varphi_{1}(x) + \ldots + p_{n} \varphi_{n}(x)$$

## Base de Lagrange et interpolation

````python
%pip install sympy
import numpy as np
import matplotlib.pyplot as plt
from sympy import symbols, simplify, expand, lambdify
from sympy.abc import x

N = 3
showlagrange = 1

# Fonction cible
def g(x_val):
    return 2.2*np.sin(x_val)**2

# Points d'interpolation
x_pts = np.linspace(0, np.pi, N)
y_pts = [g(val) for val in x_pts]

# Définition des bases de Lagrange
def lagrange_basis(j, x_sym, x_vals):
    L = 1
    x_j = x_vals[j]
    for m, x_m in enumerate(x_vals):
        if m != j:
            L *= (x_sym - x_m) / (x_j - x_m)
    return L

# Construction du polynôme de Lagrange
P = sum(y_pts[j] * lagrange_basis(j, x, x_pts) for j in range(len(x_pts)))
P_simplified = simplify(expand(P))

# Transformation en fonction utilisable avec numpy
P_func = lambdify(x, P_simplified, modules=['numpy'])

# Calcul des bases de Lagrange
L_syms = [lagrange_basis(j, x, x_pts) for j in range(len(x_pts))]
L_funcs = [lambdify(x, Lj, modules=['numpy']) for Lj in L_syms]

# Tracé
x_plot = np.linspace(0, np.pi, 400)
y_true = g(x_plot)
y_interp = P_func(x_plot)

plt.figure(figsize=(10, 6))
plt.plot(x_plot, y_true, label='$f(x) = \sin^2(x)$', color='blue')
plt.plot(x_pts, y_pts, 'ro')
for j in range(len(x_pts)):
    plt.text(x_pts[j]+0.05, y_pts[j]+0.03, f'$P_{j}$', color='r', size = 12)

plt.title("Interpolation de Lagrange de sin²(x)")
plt.xlabel("x")
plt.ylabel("y")
if showlagrange:
    for j, Lf in enumerate(L_funcs):
        plt.plot(x_plot, Lf(x_plot), alpha=0.4, label=f'$\phi_{j}(x)$')
        plt.plot(x_pts[j], 0, 'ro', alpha=0.4)
        
plt.plot(x_plot, y_interp, label='Interpolation de Lagrange', linestyle='--', color='red')

plt.legend()
plt.grid(True)
plt.show()
````
